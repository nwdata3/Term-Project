## Monte Carlo Simulation

```python
# Prep
mean_daily_returns = daily_returns.mean()         # length = n_assets
cov_matrix = daily_returns.cov()
trading_days = 252

asset_names = daily_returns.columns
n_assets = len(asset_names)

# Monte Carlo simulation
np.random.seed(42)  # reproducible
num_portfolios = 10_000
results = np.zeros((3, num_portfolios))  # rows: return, volatility, Sharpe
weights_record = []

for i in range(num_portfolios):
    weights = np.random.random(n_assets)
    weights /= weights.sum()  # long-only, sum to 1
    weights_record.append(weights)

    annual_return = np.dot(mean_daily_returns.values, weights) * trading_days
    annual_volatility = np.sqrt(weights @ (cov_matrix.values * trading_days) @ weights)
    sharpe_ratio = annual_return / annual_volatility if annual_volatility > 0 else np.nan

    results[0, i] = annual_return
    results[1, i] = annual_volatility
    results[2, i] = sharpe_ratio

results_df = pd.DataFrame({
    'Return': results[0],
    'Volatility': results[1],
    'Sharpe Ratio': results[2]
})
weights_df = pd.DataFrame(weights_record, columns=asset_names)
portfolio_df = pd.concat([results_df, weights_df], axis=1)

# Identify max Sharpe ratio portfolio
max_sharpe_idx = results_df['Sharpe Ratio'].idxmax()
optimal_port = portfolio_df.loc[max_sharpe_idx]

# Plot the portfolios
plt.figure(figsize=(12, 8))
plt.scatter(results_df['Volatility'], results_df['Return'],
            c=results_df['Sharpe Ratio'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Sharpe Ratio')
plt.scatter(optimal_port['Volatility'], optimal_port['Return'],
            c='red', marker='*', s=200, label='Optimal Portfolio')
plt.title('Monte Carlo Simulation of Portfolio Allocations')
plt.xlabel('Annualized Volatility')
plt.ylabel('Annualized Return')
plt.legend()
plt.grid(True)
plt.show()

# Show optimal portfolio weights
print("Optimal Portfolio Allocation:")
print(optimal_port)
```

```python
idx = stocks.index  # sorted DatetimeIndex

def next_valid(ts):
    ts = pd.Timestamp(ts)
    i = idx.searchsorted(ts, side="left")
    return idx[i] if i < len(idx) else idx[-1]  # clamp to last if beyond range

def prev_valid(ts):
    ts = pd.Timestamp(ts)
    i = idx.searchsorted(ts, side="right") - 1
    return idx[i] if i >= 0 else idx[0]         # clamp to first if before range

beginTrain2 = max(next_valid(beginTrain), idx[0])
endTrain2   = min(prev_valid(endTrain),   idx[-1])
beginTest2  = max(next_valid(beginTest),  idx[0])
endTest2    = min(prev_valid(endTest),    idx[-1])

train_df = stocks.loc[beginTrain2:endTrain2].copy()
test_df  = stocks.loc[beginTest2:endTest2].copy()
```

```python
import pandas as pd
import numpy as np

# --- helpers ---
def month_ends_from_index(idx: pd.DatetimeIndex) -> list[pd.Timestamp]:
    # last AVAILABLE date in each month from your data
    return (pd.Series(idx, index=idx)
            .groupby(idx.to_period("M")).max()
            .tolist())

def ann_sharpe(r):
    s = r.std(ddof=0)
    return np.nan if s == 0 else r.mean()/s*np.sqrt(252)

def backtest(weights: pd.DataFrame, rets: pd.DataFrame, cost_bps=10):
    # apply weights to NEXT day's returns (no look-ahead)
    W = weights.reindex(rets.index).ffill().shift(1).fillna(0.0)
    R = rets.reindex(W.index)
    turn = W.diff().abs().sum(axis=1).fillna(0.0)
    costs = turn * (cost_bps/1e4)
    port_ret = (W * R).sum(axis=1) - costs
    cum = (1 + port_ret).cumprod()
    mdd = (cum/cum.cummax() - 1).min()
    return {"sr": ann_sharpe(port_ret), "cum": cum, "mdd": float(mdd), "turn": float(turn.mean()), "ret": port_ret}
```

## Momentum Trading Setup

```python
def cs_momentum_weights(prices: pd.DataFrame, lookback=252, skip=21,
                        top_q=0.2, bottom_q=0.2, long_only=False) -> pd.DataFrame:
    # 12–1 momentum signal
    mom = prices.shift(skip) / prices.shift(lookback + skip) - 1
    rebal_dates = month_ends_from_index(mom.index)

    W = pd.DataFrame(index=mom.index, columns=mom.columns, dtype=float)

    for dt in rebal_dates:
        if dt not in mom.index:  # safety, but should be present now
            continue
        s = mom.loc[dt].dropna()
        if s.empty:
            continue

        n = len(s)
        k_top = max(1, int(np.floor(top_q * n)))
        top = s.nlargest(k_top).index

        if long_only:
            w = pd.Series(0.0, index=mom.columns)
            w.loc[top] = 1.0 / k_top
        else:
            k_bot = max(1, int(np.floor(bottom_q * n)))
            bot = s.nsmallest(k_bot).index
            w = pd.Series(0.0, index(mom.columns))
            w.loc[top] =  0.5 / k_top
            w.loc[bot] = -0.5 / k_bot

        W.loc[dt] = w

    return W.dropna(how="all").reindex(mom.index).ffill()
```

```python
def ts_momentum_weights(prices: pd.DataFrame, lookback=252, skip=21, long_only=False) -> pd.DataFrame:
    mom = prices.shift(skip) / prices.shift(lookback + skip) - 1
    rebal_dates = month_ends_from_index(mom.index)
    sig = mom.loc[rebal_dates].copy()

    sig = np.sign(sig)  # +1/-1 per asset
    if long_only:
        sig = sig.clip(lower=0)

    # normalize each rebalance date (sum|w|=1 long/short, sum=1 long-only)
    if long_only:
        row_sums = sig.clip(lower=0).sum(axis=1).replace(0, np.nan)
        Wm = sig.clip(lower=0).div(row_sums, axis=0).fillna(0.0)
    else:
        row_sums = sig.abs().sum(axis=1).replace(0, np.nan)
        Wm = sig.div(row_sums, axis=0).fillna(0.0)

    return Wm.reindex(prices.index).ffill()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === 0) Inputs
stocks = stocks.sort_index()
try:
    daily_returns = daily_returns.sort_index()
except NameError:
    daily_returns = stocks.ffill().pct_change().dropna(how="all").sort_index()

TEST_START = pd.Timestamp("2025-01-02")
TEST_END   = pd.Timestamp("2025-08-14")

# === 1) Helpers ===
def month_ends_from_index(idx: pd.DatetimeIndex) -> list[pd.Timestamp]:
    return (pd.Series(idx, index=idx).groupby(idx.to_period("M")).max().tolist())

def ann_sharpe(r: pd.Series) -> float:
    r = r.dropna()
    s = r.std(ddof=0)
    return np.nan if (len(r) == 0 or s == 0) else r.mean() / s * np.sqrt(252)

def backtest(weights: pd.DataFrame, rets: pd.DataFrame, cost_bps=10):
    W = weights.reindex(rets.index).ffill().shift(1).fillna(0.0)
    R = rets.reindex(W.index)

    turnover = W.diff().abs().sum(axis=1).fillna(0.0)
    costs = turnover * (cost_bps / 1e4)

    port_ret = (W * R).sum(axis=1) - costs
    cum = (1 + port_ret).cumprod()
    return {"ret": port_ret, "cum": cum, "sr": ann_sharpe(port_ret)}

# === 2) Cross-Sectional Momentum (12–1), monthly rebalance ===
def cs_momentum_weights(prices: pd.DataFrame, lookback=252, skip=21,
                        top_q=0.2, bottom_q=0.2, long_only=False) -> pd.DataFrame:
    # 12–1 momentum signal
    mom = prices.shift(skip) / prices.shift(lookback + skip) - 1
    rebal_dates = month_ends_from_index(mom.index)
    W = pd.DataFrame(index=mom.index, columns=mom.columns, dtype=float)

    for dt in rebal_dates:
        if dt not in mom.index:
            continue
        s = mom.loc[dt].dropna()
        if s.empty:
            continue

        n = len(s)
        k_top = max(1, int(np.floor(top_q * n)))
        top = s.nlargest(k_top).index

        if long_only:
            w = pd.Series(0.0, index=mom.columns, dtype=float)
            w.loc[top] = 1.0 / k_top
        else:
            k_bot = max(1, int(np.floor(bottom_q * n)))
            bot = s.nsmallest(k_bot).index
            w = pd.Series(0.0, index=mom.columns, dtype=float)
            w.loc[top] =  0.5 / k_top
            w.loc[bot] = -0.5 / k_bot

        W.loc[dt] = w

    return W.dropna(how="all").reindex(mom.index).ffill()

# Build weights and backtest
W_cs = cs_momentum_weights(stocks, lookback=252, skip=21, top_q=0.2, bottom_q=0.2, long_only=False)
bt_cs = backtest(W_cs, daily_returns.loc[TEST_START:TEST_END], cost_bps=10)

# === 3) XLE benchmark over same dates ===
try:
    bench  
except NameError:
    import yfinance as yf
    xle_df = yf.download(
        "XLE",
        start=bt_cs["ret"].index.min() - pd.Timedelta(days=5),
        end=bt_cs["ret"].index.max() + pd.Timedelta(days=1),
        progress=False,
    )
    close_like = xle_df["Adj Close"] if "Adj Close" in xle_df.columns else xle_df["Close"]
    price = close_like.squeeze().sort_index().ffill()
    bench = price.pct_change().dropna()

# Align to identical dates
port_ret = bt_cs["ret"].dropna()
bench_aligned = bench.reindex(port_ret.index).dropna()
port_ret_aligned = port_ret.reindex(bench_aligned.index)

# Metrics
port_cum  = (1 + port_ret_aligned).cumprod()
bench_cum = (1 + bench_aligned).cumprod()
port_dd   = port_cum / port_cum.cummax() - 1.0
bench_dd  = bench_cum / bench_cum.cummax() - 1.0

port_sr  = ann_sharpe(port_ret_aligned)
bench_sr = ann_sharpe(bench_aligned)

print(f"CS Momentum — Sharpe: {port_sr:.2f}, Cumulative: {port_cum.iloc[-1]:.2f}x")
print(f"XLE Benchmark — Sharpe: {bench_sr:.2f}, Cumulative: {bench_cum.iloc[-1]:.2f}x")

# === 4) Plot: cumulative & drawdown ===
fig, ax = plt.subplots(2, 1, figsize=(11, 7), sharex=True)

# Top: cumulative growth
ax[0].plot(port_cum,  label=f"QB Energy Fund (SR={port_sr:.2f}, Cum={port_cum.iloc[-1]:.2f}x)")
ax[0].plot(bench_cum, label=f"XLE (SR={bench_sr:.2f}, Cum={bench_cum.iloc[-1]:.2f}x)")
ax[0].set_title("QB Energy Fund vs XLE — Cumulative Return")
ax[0].set_ylabel("Growth of $1")
ax[0].legend(loc="best")
ax[0].grid(True, alpha=0.3)

# Bottom: drawdowns
ax[1].plot(port_dd,  label="QB Energy Fund DD")
ax[1].plot(bench_dd, label="XLE DD")
ax[1].set_title("Drawdown")
ax[1].set_ylabel("Drawdown")
ax[1].set_xlabel("Date")
ax[1].legend(loc="best")
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Optional: save
fig.savefig("cs_momentum_vs_xle.png", dpi=150, bbox_inches="tight")
print("Saved figure to cs_momentum_vs_xle.png")
```
